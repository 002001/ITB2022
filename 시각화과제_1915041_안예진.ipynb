```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ITB/health_data.csv', encoding='utf-8')
df
```

```python
#데이터 추가
df['BMI지수'] = df['체중(5Kg 단위)'] / (df['신장(5Cm단위)'] * df['신장(5Cm단위)'] / 10000)
df
```

```python
display(df.info())
```

```python
#결측치수
display(df.isna().sum())
```

```python
#결측치 처리1 - 제거

df.dropna(axis='index',how='any',subset = ['흡연상태'],inplace=True)
df.dropna(axis='index',how='any',subset = ['음주여부'],inplace=True)
df.dropna(axis='index',how='any',subset = ['트리글리세라이드'],inplace=True)
df.dropna(axis='index',how='any',subset = ['HDL콜레스테롤'],inplace=True)
df.dropna(axis='index',how='any',subset = ['LDL콜레스테롤'],inplace=True)
df
```

```python
#결측치 처리2 - 대체 (과반수 이상의 데이터가 NaN 일때)

df['치아우식증유무'].fillna('미측정',inplace=True)
df['치석'].fillna('미측정',inplace=True)
df
```

```python
#결측치 처리3 - 보간 (최빈값)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "most_frequent")
df_imputer = pd.DataFrame(imputer.fit_transform(df))
print("\n평균으로 보간:",df_imputer.shape)
print(df_imputer.isna().sum())
```

#시각화3 - 꺾은선 그래프

```python
#평균표 만들기
import numpy as np
def my_mean(s):
    return np.mean(s)

df_bmi = df.groupby('연령대코드(5세단위)').agg({'BMI지수': np.mean})
df_bmi
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

sns.lineplot(x='연령대코드(5세단위)', y="BMI지수", data=df_bmi)
sns.set(rc = {'figure.figsize':(10,7)})
```

###인사이트 도출
제공된 신장과 체중을 통해 비만도를 나타내는 'BMI지수' 항목을 추가하였다. BMI지수는 자신의 몸무게를 키의 제곱으로 나누는 것으로, BMI가 18.5 이하면 저체중 ／ 18.5 ~ 22.9 사이면 정상 ／ 23.0 ~ 24.9 사이면 과체중 ／ 25.0 이상부터는 비만으로 판정된다. 연령대별 비만율을 확인하고자 시각화를 진행하였다.

** 연령대 코드 8이하(39세 이하)와 18(85세 이상)은 다른 연령대에 비해 표본이 적어 무의미한 정보였음.

표본이 많은 40세 이상부터 85세 미만, 즉 대다수의 중장년층이 과체중을 겪고 있음을 확인할 수 있었다. 비만은 만병의 근원으로 고혈압이나 당뇨 등 만성질환의 위험에 무방비로 노출될 수 있다. 앞서 확인했던 연령대별 식전혈당 (당뇨병)과 연관지어 데이터를 분석, 시각화한다면 연관 관계를 확인할 수 있을 것이라 보인다.
